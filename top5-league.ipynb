```python
import pandas as pd 
```

```python
data = pd.read_html('https://fbref.com/en/comps/Big5/Big-5-European-Leagues-Stats', attrs={"id":"big5_table"})[0]
```

```python
data['Country'].unique()
```

```python
data['Country'] = data['Country'].replace({
    'it ITA': 'Italy', 
    'de GER': 'Germany', 
    'es ESP': 'Spain', 
    'eng ENG': 'England', 
    'fr FRA': 'France'
})

data.head()
```

```python
data['last_5_list'] = data['Last 5'].str.split()

data['last_5_wins'] = data['last_5_list'].apply(lambda x: x.count('W'))
data['last_5_loss'] = data['last_5_list'].apply(lambda x: x.count('L'))

data.head()
```

```python
data = data.drop('last_5_list', axis=1)
data.head()
```
